```python
from fed_batch_1 import (
initial_conditions,
mass_balance
)

from scipy.integrate import odeint

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Time vector
t = np.linspace(0, 25)
```

```python
# Solve ODE
x = odeint(mass_balance.mass_balance,initial_conditions.init,t, full_output=1)
```

```python
# Plot the results
plt.plot(t, x[0][:, 0], 'b', label= 'Glucose')
plt.plot(t, x[0][:, 1], 'r', label= 'Biomass')
plt.plot(t, x[0][:, 2], 'g', label= 'Volume')
plt.ylabel('concentration (g/L)')
plt.xlabel('time (h)')
plt.legend()
```
