```python
from batch import (
initial_conditions,
mass_balance
)
from scipy.integrate import odeint

import numpy as np
import matplotlib.pyplot as plt


```

```python
# Time vector
t = np.linspace(0, 30)
```

```python
# Solve ODE
x = odeint(mass_balance.mass_balance,initial_conditions.init,t)
```

```python
# Plot the results
plt.plot(t, x[:, 0], 'b', label= 'Glucose')
plt.plot(t, x[:, 1], 'r', label= 'Biomass')
plt.ylabel('concentration (g/L)')
plt.xlabel('time (h)')
plt.legend()
```
